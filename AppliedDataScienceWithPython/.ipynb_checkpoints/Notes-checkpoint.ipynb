## The Python : map() 

#### map(function, iterables). <br>
The map() function executes a specified function for each item in a iterable. The item is sent to the function as a parameter.

```python
def add(a, b):
    return a+b
```

```python
x = map(add,(1,2,3),(4,5,6))
x
```

```python
for i in x:
    print(i)
```

```python
p = [1,4.0,5]
q = [2,3,6.0]
y = map(min, p, q)

for j in y:
    print(j)
```

```python
r = ["abc", "def", "ghi"]

z = map(list, r)

for i in z:
    print(i)
    
zl = list(map(list, r))
zl
```


## The Python : Lambda and List Comprehensions
#### A lambda function can take any number of arguments, but can only have one expression.

lambda arguments : expression

```python
func = lambda a,b,c : b+c-a
```

```python
func(3,4,1)
```

The power of lambda is better shown when you use them as an anonymous function inside another function.

```python
def mulfactor(n):
    return lambda a: a*n
doubler = mulfactor(2)
tripler = mulfactor(3)
```

```python
print(doubler(10))
print(doubler(15))
print(tripler(15))
```

```python
numbers = (1,2,3,4)
z = map(lambda x: x*2, numbers)

for i in z:
    print(i)
```

```python
n1 = (1,2,3,4)
n2 = [5,6,7,8,9]

z = map(lambda x,y: x+y, n1, n2)

for i in z:
    print(i)
```

```python
my_list = []
for i in range(5):
    for j in range(5):
        my_list.append(i*j)
my_list
```

```python
my_list = [i*j for i in range(5) for j in range(5)]
my_list
```

```python
my_list_2 = ["Even" if i%2 == 0 else "odd" for i in my_list]
print(my_list_2)
```

```python
matrix = [[1,2],[3,4],[5,6]]
transpose = []
for i in range(len(matrix[0])):
    temp = []
    for row in matrix:
        temp.append(row[i])
    transpose.append(temp)

    
print(transpose)
```

```python
transpose = [[row[i] for row in matrix] for i in range(len(matrix[0]))]

print(transpose)
```

### Groupby

```python
df = pd.read_csv('census.csv')
df = df[df['SUMLEV']==50]
df.head()
```

```python
%%timeit -n 10
for state in df['STNAME'].unique():
    avg = np.average(df[df['STNAME'] == state].dropna()['CENSUS2010POP'])
    print(state + ' -> ' + str(avg))
```

```python
%%timeit -n 10
for state, frames in df.groupby('STNAME'):
    avg = np.average(frames['CENSUS2010POP'])
    print(state + ' -> ' + str(avg))
```

```python
chance_of_tornado = 0.01

tornado_events = np.random.binomial(1, chance_of_tornado, 1000000)
    
two_days_in_a_row = 0
for j in range(1,len(tornado_events)-1):
    if tornado_events[j]==1 and tornado_events[j-1]==1:
        two_days_in_a_row+=1

print('{} tornadoes back to back in {} years'.format(two_days_in_a_row, 1000000/365))
```
